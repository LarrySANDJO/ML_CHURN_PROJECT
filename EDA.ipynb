```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import KNNImputer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import IsolationForest, RandomForestClassifier
from sklearn.neighbors import LocalOutlierFactor
from sklearn.covariance import EllipticEnvelope
#from imblearn.over_sampling import RandomOverSampler
from sklearn.svm import SVC, OneClassSVM
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.decomposition import PCA

from time import perf_counter

import warnings
warnings.filterwarnings('ignore')
```

```python
class ExtraServicesCounter(BaseEstimator, TransformerMixin):
    def __init__(self, service_cols):
        self.service_cols = service_cols

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        X['total_extra_services'] = X[self.service_cols].applymap(lambda val: 1 if val == 'Yes' else 0).sum(axis=1)
        return X.drop(columns=self.service_cols)

class MonthlyFeaturesCreator(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        X['New_monthly_charges'] = X['Total Charges'] / X['Tenure in Months'].replace(0, np.nan)
        X['Ratio_refund'] = X['Total Refunds'] / (X['Tenure in Months'] * X['New_monthly_charges']).replace(0, np.nan)
        X['Extra_charges_ratio'] = (X['Total Extra Data Charges'] + X['Total Long Distance Charges']) / X['Total Charges'].replace(0, np.nan)
        X['Avg Monthly GB Download'] = X['Avg Monthly GB Download'].fillna(0)
        return X.drop(columns=['Monthly Charge', 'Total Charges', 'Total Extra Data Charges', 'Total Long Distance Charges', 'Total Refunds', 
                               'Latitude', 'Longitude', 'Avg Monthly Long Distance Charges', 'Paperless Billing', 'Payment Method'])

class OutlierIQRImputer(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        self.bounds = {}
        for col in self.columns:
            q1 = X[col].quantile(0.25)
            q3 = X[col].quantile(0.75)
            iqr = q3 - q1
            lower = q1 - 1.5 * iqr
            upper = q3 + 1.5 * iqr
            self.bounds[col] = (lower, upper)
        return self

    def transform(self, X):
        X = X.copy()
        for col in self.columns:
            lower, upper = self.bounds[col]
            X[col] = X[col].clip(lower, upper)
        return X

```

```python
# Étapes
feature_engineering = Pipeline(steps=[
    ("extra_services", ExtraServicesCounter(service_cols=[
        'Online Security', 'Online Backup', 'Device Protection Plan',
        'Premium Tech Support', 'Streaming TV', 'Streaming Movies', 'Streaming Music', 'Unlimited Data'
    ])),
    ("monthly_features", MonthlyFeaturesCreator()),
])

# Colonnes restantes
numerical_cols = ['Avg Monthly GB Download', 'Tenure in Months', 'Number of Referrals', 'Number of Dependents', 'Age',
                  'New_monthly_charges', 'Ratio_refund', 'Extra_charges_ratio', 'total_extra_services']
categorical_cols = ['Internet Service', 'Phone Service', 'Multiple Lines', 'Married', 'Gender']
freq_encoded_cols = ['Offer', 'Internet Type', 'City']
ordinal_col = ['Contract']

# Encodages
preprocessing = ColumnTransformer(transformers=[
    ("num", Pipeline(steps=[
        ("imputer", SimpleImputer(strategy='median')),
        ("scaler", MinMaxScaler())
    ]), numerical_cols),

    ("onehot", OneHotEncoder(handle_unknown='ignore'), categorical_cols),

    ("freq_encode", "passthrough", freq_encoded_cols), 

    ("contract_ord", OrdinalEncoder(categories=[['Month-to-Month', 'One Year', 'Two Year']]), ordinal_col)
])
preprocessing
```

```python
pipeline = Pipeline(steps=[
    ("feature_engineering", feature_engineering),
    ("preprocessing", preprocessing),
    ("model", XGBClassifier(eval_metric='logloss', use_label_encoder=False))
])
```

```python
# ========================
# Custom Transformers
# ========================

class ExtraServicesTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, service_cols):
        self.service_cols = service_cols

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        X['total_extra_services'] = X[self.service_cols].apply(lambda row: sum([1 if val == 'Yes' else 0 for val in row]), axis=1)
        return X.drop(columns=self.service_cols)

class FeatureEngineer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()

        # Replace NA in Multiple Lines
        X['Multiple Lines'] = X['Multiple Lines'].fillna('No')

        # Avg Monthly GB Download NA -> 0
        X['Avg Monthly GB Download'] = X['Avg Monthly GB Download'].fillna(0)

        # New Monthly Charge
        X['New_monthly_charge'] = X['Total Charges'] / X['Tenure in Months']
        X['New_monthly_charge'].replace([np.inf, -np.inf], np.nan, inplace=True)
        
        # Ratio Refund
        denom = X['Tenure in Months'] * X['New_monthly_charge']
        X['Ratio_refund'] = X['Total Refunds'] / denom
        X['Ratio_refund'].replace([np.inf, -np.inf], np.nan, inplace=True)

        # Extra charges ratio
        extra_charges = X['Total Extra Data Charges'] + X['Total Long Distance Charges']
        X['Extra_charges_ratio'] = extra_charges / X['Total Charges']
        X['Extra_charges_ratio'].replace([np.inf, -np.inf], np.nan, inplace=True)

        # Drop specified columns
        drop_cols = [
            'Total Charges', 'Total Refunds', 'Total Extra Data Charges',
            'Total Long Distance Charges', 'Monthly Charge', 'Total Revenue',
            'Zip Code', 'Latitude', 'Longitude', 'CustomerID', 'Churn Category',
            'Payment Method', 'Paperless Billing'
        ]
        X = X.drop(columns=drop_cols, errors='ignore')
        
        return X

class FrequencyEncoder(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns
        self.frequencies = {}

    def fit(self, X, y=None):
        for col in self.columns:
            freq = X[col].value_counts(normalize=True)
            self.frequencies[col] = freq.to_dict()
        return self

    def transform(self, X):
        X = X.copy()
        for col in self.columns:
            freq_map = self.frequencies.get(col, {})
            # map values, set default freq to 0 if unseen category
            X[col] = X[col].map(freq_map).fillna(0)
        return X


class OutlierClipper(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        self.bounds = {}
        for col in self.columns:
            q1 = X[col].quantile(0.25)
            q3 = X[col].quantile(0.75)
            iqr = q3 - q1
            lower = q1 - 1.5 * iqr
            upper = q3 + 1.5 * iqr
            self.bounds[col] = (lower, upper)
        return self

    def transform(self, X):
        X = X.copy()
        for col, (lower, upper) in self.bounds.items():
            X[col] = np.clip(X[col], lower, upper)
        return X

# ========================
# Pipeline Construction
# ========================

service_cols = [
    'Online Security', 'Online Backup', 'Device Protection Plan',
    'Premium Tech Support', 'Streaming TV', 'Streaming Movies', 'Streaming Music', 'Unlimited Data'
]

onehot_cols = ['Internet Service', 'Phone Service', 'Multiple Lines', 'Married']
ordinal_cols = ['Contract']
freq_enc_cols = ['Offer', 'Internet Type', 'City']
numeric_cols = ['Avg Monthly GB Download', 'Tenure in Months', 'Number of Referrals',
                'Number of Dependents', 'Age', 'New_monthly_charge',
                'Ratio_refund', 'Extra_charges_ratio', 'total_extra_services']

pipeline = Pipeline(steps=[
    ('extra_services', ExtraServicesTransformer(service_cols=service_cols)),
    ('feature_engineering', FeatureEngineer()),
    ('frequency_encoding', FrequencyEncoder(columns=freq_enc_cols)),
    ('outlier_clipping', OutlierClipper(columns=numeric_cols)),
    ('preprocessor', ColumnTransformer(transformers=[
        ('num', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', MinMaxScaler())
        ]), numeric_cols),
        ('onehot', OneHotEncoder(handle_unknown='ignore'), onehot_cols),
        ('ordinal', OrdinalEncoder(categories=[['Month-to-Month', 'One Year', 'Two Year']]), ordinal_cols)
    ], remainder='drop')),
    ('clf', XGBClassifier(random_state=42, eval_metric='logloss'))
])
pipeline
```

```python
class IQRTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        # Si X est un array, le transformer en DataFrame
        if isinstance(X, pd.DataFrame):
            df = X
        else:
            df = pd.DataFrame(X, columns=self.columns)
        
        self.bounds_ = {}
        for col in self.columns:
            q1 = df[col].quantile(0.25)
            q3 = df[col].quantile(0.75)
            iqr = q3 - q1
            lower = q1 - 1.5 * iqr
            upper = q3 + 1.5 * iqr
            self.bounds_[col] = (lower, upper)
        return self

    def transform(self, X):
        if isinstance(X, pd.DataFrame):
            df = X.copy()
        else:
            df = pd.DataFrame(X, columns=self.columns)

        for col in self.columns:
            lower, upper = self.bounds_[col]
            df[col] = df[col].clip(lower, upper)
        return df
```
